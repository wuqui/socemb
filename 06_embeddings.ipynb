# Embeddings

## Imports

```python
from flair.embeddings import WordEmbeddings
from flair.data import Sentence
```

```python
from flair.embeddings import TransformerWordEmbeddings
```

## Type-based embeddings

```python
embedding = WordEmbeddings('glove')
```

```python
sentence = Sentence('I can can a can.')
embedding.embed(sentence)
```

```python
for token in sentence:
    print(token)
    print(token.embedding.shape)
```

# Token-based embeddings

```python
embedding = TransformerWordEmbeddings('roberta-base')
```

```python
sentence = Sentence("It's okay to use the term Anglo-Saxon.")
embs = embedding.embed(sentence)
```

```python
for tok in sentence:
    if tok.text == 'Anglo-Saxon':
        print(tok, tok.embedding.shape)
```
